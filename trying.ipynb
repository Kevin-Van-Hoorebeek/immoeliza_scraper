```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.service import Service
import time
import re

root_url = "https://www.immoweb.be/"
Listings_url = root_url + 'en/search/house-and-apartment/for-sale'

# Specify the path to your chromedriver
service = Service(executable_path=r"C:\Users\Fox\Documents\GitHub\immoeliza_scraper\chromedriver-win64\chromedriver.exe")
driver = webdriver.Chrome(service=service)

try:
    # Open the website
    driver.get( Listings_url)

    # Wait for the GDPR pop-up to appear and accept it
    shadow_host = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "#usercentrics-root"))
    )

    # Access the shadow root
    shadow_root = driver.execute_script('return arguments[0].shadowRoot', shadow_host)

    # Wait for and click the accept button
    accept_button = WebDriverWait(driver, 10).until(
        lambda d: shadow_root.find_element(By.CSS_SELECTOR, "button[data-testid='uc-accept-all-button']")
    )
    accept_button.click()

    # Wait for the button to disappear
    WebDriverWait(driver, 10).until(EC.staleness_of(accept_button))

    # Find and click the search button
    search_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.ID, 'searchBoxSubmitButton'))
    )
    search_button.click()

    # Optional: Wait for some time to see results (if needed)
    time.sleep(15)
    

```

```python
# Scraping with Selenium
    # make a list
property_dataset = []

def scrape_page():
    # Find all search result items (update the selector based on the actual site's HTML structure)
    results = driver.find_elements(By.CSS_SELECTOR, '."search-results__list"')

    for result in results:
        link = result.find_element(By.TAG_NAME, 'a').get_attribute('href')  # Assuming the link is an anchor tag
        print(f"Scraping: {link}")

        
        # Navigate to the item page
        driver.get(link)
        
        # Optional: wait for the page to load
        time.sleep(3)

        # Scrape the desired information

        property_type = re.search(r'appartement | house', link)
        # if match:
        #     print(match.group())
        property_sub_type = driver.find_element(By.CSS_SELECTOR, '.classified__title').text
        property_description = driver.find_element(By.CSS_SELECTOR, '.text-block__body read-toggle__container--expanded').text
        property_ID = driver.find_element(By.CSS_SELECTOR, '.classified__header--immoweb-code').text
        # property_specs = driver.find_element(By.CSS_SELECTOR, '.classified__header--immoweb-code').text
        property_code = driver.find_element(By.CLASS_NAME, 'classified__information--address-row').text.split()
        # property_location = driver.find_element_by_xpath('.//*[@id="classified-header"]/div/div/div[1]/div/div[2]/div[1]/div/button/span[2]/text()[2]').text
        
        # Print or store the scraped information
        property_attris = {
                'Property type': property_type,
                'Property subtype': property_sub_type,
                'Property ID': property_ID,
                
        

        } 

        property_dataset.append(property_attris)

        # Navigate back to the search results page
        driver.back()
        
        # Optional: wait for the page to load
        time.sleep(3)


# Loop through all pagination pages
while True:
    scrape_page()
    
    # Check if there's a next page
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, '.pagination__link pagination__link--next button button--text button--size-small')
        if next_button.is_enabled():
            next_button.click()  # Click the next page button
            time.sleep(3)  # Wait for the page to load
        else:
            break  # Exit loop if next button is not enabled
    except Exception as e:
        print("No more pages or an error occurred:", e)
        break

# Close the driver
finally:
    driver.quit()
```
